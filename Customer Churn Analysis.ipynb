Preprocessing of Data before Model Implementation :

```python
df['Churned']=df['Churned'].astype(int)
df['IsAutoRenew']=df['IsAutoRenew'].astype(int)
```

```python
df
```

```python
categorical_features = df.select_dtypes(include=['object', 'bool']).columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
categorical_features
numerical_features
```

```python
categorical_features
```

```python
df
```

```python
df=df.drop(['ChurnDate','Date'],axis=1)
```

```python
df
```

```python
X=df.drop(['Churned'],axis=1)
X
```

```python
y=df['Churned']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_test
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train['AgeGroup'].unique()
```

```python
oe=OrdinalEncoder(categories=[['Basic', 'Standard', 'Premium'],[ 'Annually', 'Quarterly','Monthly'],['18-25','26-35','36-50','51+']])
```

```python
oe.fit(X_train[['SubscriptionPlan','BillingCycle','AgeGroup']])
```

```python
X_test[['SubscriptionPlan','BillingCycle','AgeGroup']]=oe.transform(X_test[['SubscriptionPlan','BillingCycle','AgeGroup']])
```

```python
X_train[['SubscriptionPlan','BillingCycle','AgeGroup']]=oe.transform(X_train[['SubscriptionPlan','BillingCycle','AgeGroup']])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False,dtype='int64').set_output(transform='pandas')
```

```python
ohe_features = ['Location', 'ContentPreference', 'PrimaryDevice']
X_train_new=ohe.fit_transform(X_train[ohe_features])
X_train_new
X_test_new=ohe.transform(X_test[ohe_features])
```

```python
X_train_final=pd.concat([X_train,X_train_new],axis=1).drop(['ContentPreference','Location','PrimaryDevice'],axis=1)
```

```python
X_test_final=pd.concat([X_test,X_test_new],axis=1).drop(['ContentPreference','Location','PrimaryDevice'],axis=1)
```

```python
X_test_final
```

```python
numerical_features=numerical_features.drop('Churned')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
St=StandardScaler().set_output(transform='pandas')
```

```python
St.fit(X_train_final[numerical_features])
```

```python
X_train_1=St.transform(X_train_final[numerical_features])
```

```python
X_test_1=St.transform(X_test_final[numerical_features])
```

```python
X_test=pd.concat([X_test_final.drop(numerical_features,axis=1),X_test_1],axis=1)
```

```python
X_train=pd.concat([X_train_final.drop(numerical_features,axis=1),X_train_1],axis=1)
```
